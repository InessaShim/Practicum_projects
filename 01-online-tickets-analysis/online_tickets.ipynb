```python
# Фильтрация выбросов по 99‑му процентилю для revenue:
rub_revenue_99 = df_clean[df_clean['currency_code'] == 'rub']['revenue'].quantile(0.99)
kzt_revenue_99 = df_clean[df_clean['currency_code'] == 'kzt']['revenue'].quantile(0.99)

print(f"99-й процентиль revenue (RUB): {rub_revenue_99}")
print(f"99-й процентиль revenue (KZT): {kzt_revenue_99}")

# Фильтруем данные: 99‑й процентиль + отрицательные значения
df_filtered = df_clean[
    (
        ((df_clean['currency_code'] == 'rub') & (df_clean['revenue'] <= rub_revenue_99) & (df_clean['revenue'] >= 0)) |
        ((df_clean['currency_code'] == 'kzt') & (df_clean['revenue'] <= kzt_revenue_99) & (df_clean['revenue'] >= 0))
    )
].copy()

print(f"Удалено записей из‑за выбросов и отрицательных значений: {len(df_clean) - len(df_filtered)}")
```

```python
# Преобразование типов данных

# Конвертация дат
df_filtered['created_dt_msk'] = pd.to_datetime(df_filtered['created_dt_msk'])
df_filtered['created_ts_msk'] = pd.to_datetime(df_filtered['created_ts_msk'])
```

```python
# Оптимизация типов данных
# Категориальные данные
categorical_columns = ['currency_code', 'device_type_canonical', 'event_type_main', 'region_name', 'city_name', 'age_limit']
for col in categorical_columns:
    df_filtered[col] = df_filtered[col].astype('category')
```

```python
# Целочисленные данные

# До преобразования заполним пропуски в столбце days_since_prev
df_filtered['days_since_prev'] = df_filtered['days_since_prev'].fillna(0)

# Список столбцов для оптимизации
int_columns = ['order_id', 'event_id', 'tickets_count', 'city_id', 'venue_id', 'days_since_prev']

# Оптимизация в цикле
for col in int_columns:
    if col in df_filtered.columns:
        df_filtered[col] = pd.to_numeric(df_filtered[col], downcast='integer')
        print(f"Столбец '{col}' оптимизирован. Новый тип: {df_filtered[col].dtype}")
```

```python
# Данные с плавающей точкой

# Список столбцов для оптимизации
float_columns = ['revenue', 'total']

# Оптимизация в цикле
for col in float_columns:
    if col in df_filtered.columns:
        df_filtered[col] = pd.to_numeric(df_filtered[col], downcast='float')
        print(f"Столбец '{col}' оптимизирован. Новый тип: {df_filtered[col].dtype}")
```

```python
# Проверим результат
df_filtered.info()
```

Типы данных оптимизированы.

```python
# Конвертация валюты в рубли

# Приводим даты курса к формату datetime
tenge_df['data'] = pd.to_datetime(tenge_df['data'])


# Объединяем с курсом
df_merged = df_filtered.merge(
    tenge_df[['data', 'curs']],
    left_on='created_dt_msk',
    right_on='data',
    how='left'
)

# Конвертируем revenue в рубли
df_merged['revenue_rub'] = np.where(
    df_merged['currency_code'] == 'rub',
    df_merged['revenue'],
    df_merged['revenue'] * (df_merged['curs'] / 100)  # курс для 100 тенге
)
```

```python
# Создаем новые столбцы

# Выручка с продажи одного билета
df_merged['one_ticket_revenue_rub'] = df_merged['revenue_rub'] / df_merged['tickets_count']


# Месяц заказа
df_merged['month'] = df_merged['created_dt_msk'].dt.month

# Сезон
def get_season(month):
    if month in [12, 1, 2]:
        return 'зима'
    elif month in [3, 4, 5]:
        return 'весна'
    elif month in [6, 7, 8]:
        return 'лето'
    else:
        return 'осень'

df_merged['season'] = df_merged['month'].apply(get_season)
```

```python
# Финальная проверка данных

print("Финальный размер датасета:", df_merged.shape)
print("\nПропуски после предобработки:")
print(df_merged.isnull().sum())


print("\nПервые 5 строк финального датасета:")
df_merged.head()
```

```python
# Распределение заказов для осеннего и летнего периода по разным категориям
# Определяем сегменты для анализа
segments = ['event_type_main', 'device_type_canonical', 'age_limit']

# Создаём словарь results с данными для визуализации
results = {}
for segment in segments:
    grouped = df_merged.groupby('season')[segment]
    # Абсолютные значения
    abs_vals = grouped.value_counts()
    # Доли
    share_vals = grouped.value_counts(normalize=True)
    # Объединяем в датафрейм
    temp_df = pd.DataFrame({'abs': abs_vals, 'share': share_vals})
    results[segment] = temp_df

# Вывод
for seg, data in results.items():
    print(f"\n--- {seg} (groupby) ---")
    print(data)
```

```python
# Визуализация
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Распределение заказов по сегментам: лето, осень', fontsize=16)

for i, segment in enumerate(segments):
    data = results[segment]['share']  
    summer = data.loc['лето']
    autumn = data.loc['осень']
    
    x = np.arange(len(summer))
    width = 0.35
    
    axes[i].bar(x - width/2, summer.values, width, label='Лето', alpha=0.8)
    axes[i].bar(x + width/2, autumn.values, width, label='Осень', alpha=0.8)
    
    axes[i].set_title(segment)
    axes[i].set_xticks(x)
    axes[i].set_xticklabels(summer.index, rotation=45, ha='right')
    axes[i].legend()
    axes[i].grid(axis='y', alpha=0.3)
    axes[i].set_ylabel('Доля')

plt.tight_layout()
plt.show()
```

Сезонность влияет на предпочтения аудитории: осенью растёт спрос на стендап и спорт, снижается доля концертов.

Мобильная аудитория стабильна: ~ 80 % заказов идёт с мобильных устройств, но осенью немного растёт доля десктопов.

Семейные и подростковые мероприятия становятся популярнее осенью: растёт доля категорий 6+ и 12+.

Мероприятия для взрослой аудитории теряют долю осенью: категория 18+ снижается с 15 % до 10 %.

###### 4. Статистический анализ данных

```python
# Разделяем пользователей по типу устройства
mobile_users = df_autumn[df_autumn['device_type_canonical'] == 'mobile']
desktop_users = df_autumn[df_autumn['device_type_canonical'] == 'desktop']
```

```python
# Сравнение размеров групп
print(f"\nРазмер группы пользователей мобильных устройств: {len(mobile_users)} пользователей")
print(f"Размер группы пользователей стационарных устройств: {len(desktop_users)} пользователей")
```

```python
# Базовая статистика по количеству заказов на пользователя
mobile_orders_stats = mobile_users.groupby('user_id')['order_id'].count()
desktop_orders_stats = desktop_users.groupby('user_id')['order_id'].count()

print("Статистика по количеству заказов на пользователя:")
print("Мобильные устройства:")
print(mobile_orders_stats.describe())
print("\nСтационарные устройства:")
print(desktop_orders_stats.describe())

# Статистика по времени между заказами
mobile_time_stats = mobile_users['days_since_prev'].dropna()
desktop_time_stats = desktop_users['days_since_prev'].dropna()

print("\nСтатистика по времени между заказами (дни):")
print("Мобильные устройства:")
print(mobile_time_stats.describe())
print("\nСтационарные устройства:")
print(desktop_time_stats.describe())
```

```python
# Визуализация распределения
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Распределение количества заказов
axes[0].hist(mobile_orders_stats, alpha=0.7, label='Мобильные', bins=20)
axes[0].hist(desktop_orders_stats, alpha=0.7, label='Стационарные', bins=20)
axes[0].set_title('Распределение количества заказов на пользователя')
axes[0].set_xlabel('Количество заказов')
axes[0].set_ylabel('Частота')
axes[0].legend()

# Распределение времени между заказами
axes[1].hist(mobile_time_stats, alpha=0.7, label='Мобильные', bins=30)
axes[1].hist(desktop_time_stats, alpha=0.7, label='Стационарные', bins=30)
axes[1].set_title('Распределение времени между заказами')
axes[1].set_xlabel('Дни с предыдущей покупки')
axes[1].set_ylabel('Частота')
axes[1].legend()

plt.tight_layout()
plt.show()
```

**Гипотеза 1. Среднее количество заказов на пользователя**

Формулировка гипотез:

H₀ (нулевая): μ mobile = μ desktop (среднее количество заказов одинаково для мобильных и стационарных устройств).

H₁ (альтернативная): μ mobile > μ desktop (среднее количество заказов выше для мобильных устройств).

```python
# Распределение данных не нормально - используем тест Манна-Уитни
alpha = 0.05
stat, p_value = stats.mannwhitneyu(mobile_orders_stats, desktop_orders_stats, alternative='greater')

# Вывод результатов
print(f"\nРезультаты статистического теста")
print(f"Выбранный тест: U‑тест Манна‑Уитни")
print(f"Уровень значимости (alpha): {alpha}")
print(f"p‑value: {p_value:.6f}")

# Интерпретация результатов
if p_value < alpha:
    conclusion = "Отвергаем нулевую гипотезу. Среднее количество заказов выше у пользователей мобильных устройств, и это различие статистически значимо."
else:
    conclusion = "Не можем отвергнуть нулевую гипотезу. Нет статистически значимых доказательств того, что среднее количество заказов пользователей мобильных устройств выше чем пользователей стационарных устройств."

print(conclusion)

# Средние значения для сравнения
mean_orders_mobile = mobile_orders_stats.mean()
mean_orders_desktop = desktop_orders_stats.mean()
print(f"\nСреднее количество заказов пользователей мобильных устройств: {mean_orders_mobile:.2f}")
print(f"Среднее количество заказов пользователей стационарных устройств:: {mean_orders_desktop:.2f}")
print(f"Разница в средних: {mean_orders_mobile - mean_orders_desktop:.2f}")
```

**Гипотеза 2. Среднее время между заказами**

Формулировка гипотез:

H₀ (нулевая): μ mobile = μ desktop (среднее время между заказами одинаково для мобильных и стационарных устройств).

H₁ (альтернативная): μ mobile > μ desktop (среднее время между заказами больше для мобильных устройств).

```python
# Распределение данных не нормально - используем тест Манна-Уитни
alpha = 0.05
stat, p_value = stats.mannwhitneyu(mobile_time_stats, desktop_time_stats, alternative='greater')

# Вывод результатов
print(f"\nРезультаты статистического теста")
print(f"Выбранный тест: U‑тест Манна‑Уитни")
print(f"Уровень значимости (alpha): {alpha}")
print(f"p‑value: {p_value:.6f}")

# Интерпретация результатов
if p_value < alpha:
    conclusion = "Отвергаем нулевую гипотезу. Среднее время между заказами выше у пользователей мобильных устройств, и это различие статистически значимо."
else:
    conclusion = "Не можем отвергнуть нулевую гипотезу. Нет статистически значимых доказательств того, что среднее время между заказами пользователей мобильных устройств выше чем пользователей стационарных устройств."

print(conclusion)

# Средние значения для сравнения
mean_time_mobile = mobile_time_stats.mean()
mean_time_desktop = desktop_time_stats.mean()
print(f"\nСреднее время между заказами пользователей мобильных устройств: {mean_time_mobile:.2f} дней")
print(f"Среднее время между заказами пользователей стационарных устройств:: {mean_time_desktop:.2f} дней")
print(f"Разница в средних: {mean_time_mobile - mean_time_desktop:.2f}")
```
